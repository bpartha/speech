```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate a time axis (for 1 second at 44100 samples/sec)
fs = 44100  # sampling rate (Hz)
t = np.linspace(0, 1, fs, endpoint=False)

# 2. Create a signal = sum of two sine waves (A4 = 440 Hz and A5 = 880 Hz)
f1, f2 = 440, 880
signal = np.sin(2 * np.pi * f1 * t) + 0.5 * np.sin(2 * np.pi * f2 * t)

# 3. Plot the signal in the time domain
plt.figure(figsize=(12, 4))
plt.title("Time Domain: Sound Wave (440 Hz + 880 Hz)")
plt.plot(t[:1000], signal[:1000])  # zoom into first 1000 samples
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.show()
```

```python
# 4. Apply Fourier Transform (FFT)
fft = np.fft.fft(signal)
freqs = np.fft.fftfreq(len(fft), 1/fs)

# Take only the positive half of the spectrum
positive_freqs = freqs[:len(freqs)//2]
magnitude = np.abs(fft[:len(freqs)//2])

# 5. Plot frequency domain
plt.figure(figsize=(12, 4))
plt.title("Frequency Domain: Spectrum of the Sound")
plt.plot(positive_freqs, magnitude)
plt.xlim(0, 2000)  # show up to 2 kHz
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.show()
```
